```python
class LGBBoosterWrapper:
    def __init__(self, booster):
        self.booster = booster

    def predict_proba(self, X):
        """
        For binary classification, the Booster's predict() usually returns
        a 1D array of positive-class probabilities. We convert it to Nx2.
        """
        p_pos = self.booster.predict(X)  # Nx1
        p_neg = 1.0 - p_pos
        # shape into Nx2
        return np.vstack([p_neg, p_pos]).T

    # Optional: also implement predict() for completeness
    def predict(self, X):
        """If needed: threshold at 0.5 for binary classification"""
        return (self.booster.predict(X) > 0.5).astype(int)

# Then pass this wrapper into DiCE:
wrapped_model = LGBBoosterWrapper(model)
```

```python
# We can wrap the LightGBM model in a DiCE model
m = dice_ml.Model(
    model=wrapped_model,
    backend="sklearn",  # or "lightgbm" if using dice-ml built-in
    model_type='classifier'
)

exp = Dice(d, m, method="random")  # or "kdtree"/"genetic"/etc
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

########################################
# 1) Define helper functions
########################################

def rescale_values(df_scaled, scaler, columns):
    """
    Rescale certain columns of df_scaled back to their original scale
    using the provided scaler. We assume the same columns that were scaled.
    """
    df_rescaled = df_scaled.copy()
    df_rescaled[columns] = scaler.inverse_transform(df_scaled[columns])
    return df_rescaled


def generate_insights_for_defense(test_data, model, scaler, cols_to_stand, total_cfs=5):
    """
    Generates counterfactuals for each instance in test_data (predicted as 0)
    that aim to flip the prediction to 1, using a DiCE explainer `exp`.

    Returns a list of dictionaries, each containing:
    - index: original index in test_data
    - original: DataFrame row of the original (rescaled)
    - cf: DataFrame of counterfactuals (rescaled)
    - feature_changes: Series of average (signed) changes across all CFs
    """
    actionable_insights = []

    for idx, instance in test_data.iterrows():
        # Reshape instance to match the model's expected input format
        instance_reshaped = instance.to_frame().T

        # Generate counterfactuals using DiCE
        dice_exp = exp.generate_counterfactuals(
            instance_reshaped,
            total_CFs=total_cfs,
            desired_class=1  # flip from 0 to 1
        )

        # The CF DataFrame (scaled form)
        cf_df = dice_exp.cf_examples_list[0].final_cfs_df

        # Rescale the CF DataFrame to original scale
        cf_df_rescaled = rescale_values(cf_df, scaler, cols_to_stand)

        # Also get the original row in rescaled form for direct comparison
        original_row_rescaled = rescale_values(instance_reshaped, scaler, cols_to_stand)

        # Compute the mean (signed) change for each feature across all CFs
        feature_changes = cf_df_rescaled.mean(axis=0) - original_row_rescaled.mean(axis=0)

        # Collect and store these insights
        actionable_insights.append({
            "index": idx,
            "original": original_row_rescaled,  # shape (1, n_features)
            "cf": cf_df_rescaled,               # shape (total_cfs, n_features)
            "feature_changes": feature_changes
        })

        # (Optionally) stop if we have enough examples
        if len(actionable_insights) >= 100:
            break

    return actionable_insights


def summarize_general_insights(actionable_insights):
    """
    Aggregates the mean (signed) feature changes across all instances and CFs.
    This is the same logic you had originally, just using the `feature_changes`
    stored in actionable_insights.
    """
    all_changes = []
    for entry in actionable_insights:
        # feature_changes is already a Series of mean changes across CFs
        all_changes.append(entry["feature_changes"])

    # Combine into a single DataFrame
    changes_df = pd.DataFrame(all_changes).fillna(0)

    # Compute the mean across all instances (signed mean)
    aggregated_changes = changes_df.mean(axis=0).sort_values(ascending=False)

    return aggregated_changes


def plot_general_insights(general_insights):
    """
    Plots the mean (signed) feature changes for each feature.
    """
    plt.figure(figsize=(12, 6))
    general_insights.plot(kind='bar', color='skyblue')
    plt.title('Mean Feature Changes', fontsize=16)
    plt.xlabel('Features', fontsize=14)
    plt.ylabel('Mean Change', fontsize=14)
    plt.xticks(rotation=45, ha='right', fontsize=12)
    plt.tight_layout()
    plt.show()


def compute_mean_abs_change(actionable_insights):
    """
    Computes the mean absolute change for each feature, considering only
    cases where the feature actually changed (non-zero difference).
    """
    # Dictionary: feature -> list of absolute differences
    feature_differences = {}

    for entry in actionable_insights:
        original = entry["original"].iloc[0]  # a Series with original row
        cf_df = entry["cf"]                  # DataFrame of CFs

        # Compute absolute differences: shape = (total_cfs, n_features)
        abs_diff = (cf_df - original).abs()

        # Accumulate non-zero differences for each feature
        for feature in abs_diff.columns:
            if feature not in feature_differences:
                feature_differences[feature] = []
            # Extend with non-zero differences
            non_zero_vals = abs_diff[feature][abs_diff[feature] != 0].tolist()
            feature_differences[feature].extend(non_zero_vals)

    # Now compute the mean of non-zero diffs per feature
    mean_changes = {}
    for feature, diffs in feature_differences.items():
        if len(diffs) > 0:
            mean_changes[feature] = np.mean(diffs)
        else:
            mean_changes[feature] = 0.0

    # Convert dict to Series, sort descending
    mean_changes_series = pd.Series(mean_changes).sort_values(ascending=False)
    return mean_changes_series


########################################
# 2) Main flow (example usage)
########################################

# --- Step A: Predict with your trained model on X_test
y_pred = model.predict(X_test)

# --- Step B: Filter only the samples the model predicts as 0
test_data_0 = X_test[y_pred < 0.5]

# --- Step C: (Optional) sample 100
sample_test_data = test_data_0.sample(n=50, random_state=42)

# --- Step D: Generate CFs with desired_class=1, flipping from 0 -> 1
insights = generate_insights_for_defense(
    test_data=sample_test_data,
    model=model,
    scaler=scaler,
    cols_to_stand=cols_to_stand,
    total_cfs=5
)

# --- Step E: Summarize (Signed) Mean Changes
general_insights = summarize_general_insights(insights)

# Print them
print("General Insights for Flipping from 0 to 1 (Signed Mean Changes):")
for feature, change_value in general_insights.items():
    direction = "increase" if change_value > 0 else "decrease"
    print(f"  Feature '{feature}': Mean change {change_value:.2f} ({direction}).")

# --- Step F: Plot the signed changes
plot_general_insights(general_insights)

# --- Step G: Compute Mean Absolute Change (only where feature changed)
mean_abs_change_series = compute_mean_abs_change(insights)

print("\nMean Absolute Change (Features that actually changed):")
print(mean_abs_change_series)

# --- Step H: Plot the mean absolute changes
plt.figure(figsize=(12, 6))
mean_abs_change_series.plot(kind='bar', color='orange')
plt.title('Mean Absolute Change (Only Non-Zero Diffs)', fontsize=16)
plt.xlabel('Features', fontsize=14)
plt.ylabel('Mean Absolute Change', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.tight_layout()
plt.show()


########################################
# 3) Example: Focused interpretation
########################################
print("\nKey Observations:")
print("1. We only generated CFs for examples originally predicted as 0.")
print("2. The signed mean changes show overall direction of push.")
print("3. The mean absolute changes show how far features move when they do change.")
```

```python
def visualize_single_play(row, row_index=None):
    """
    Given a single 'row' from your 'positions' DataFrame (merged with final_df info),
    plot the positions of players on an NFL field using columns like:
        x_WR_1.0_Possessive, y_WR_1.0_Possessive,
        x_CB_1.0_Opponent,   y_CB_1.0_Opponent,
    etc.

    row: a pandas Series containing the data of one play.
    row_index: optional integer or string label for the row (for labeling).
    """

    fig, ax = plt.subplots(figsize=(12, 6))
    draw_nfl_field(ax)  # draw the 120 x 53.3 yard field

    # We'll keep track of labels we've used so we don't get duplicates in the legend
    used_labels = set()

    # Loop over columns that start with 'x_'
    for col in row.index:
        if col.startswith("x_"):
            x_val = row[col]
            # Construct the matching y-column name
            y_col = "y_" + col[2:]  # replace 'x_' with 'y_'
            if y_col not in row.index:
                continue  # skip if there's no matching y-col

            y_val = row[y_col]

            # You might want to skip plotting if x_val or y_val are 0 or NaN
            if pd.isna(x_val) or pd.isna(y_val):
                continue

            # Extract role and team from the column name: x_<ROLE>_<N>_<TEAM>
            parts = col.split('_')
            if len(parts) < 4:
                continue  # unexpected format, skip

            role = parts[1]
            number = parts[2]
            team_label = parts[3]

            # Decide color or marker:
            if "Opponent" in team_label:
                color = "red"
                marker = "o"
            else:
                color = "blue"
                marker = "o"

            # Special highlight for the QB
            if role.upper() == "QB":
                marker = "s"   # square for QB
                color = "green"

            # Create a label that might appear in the legend
            if "Opponent" in team_label:
                legend_label = f"{role} (Def)"
            else:
                legend_label = f"{role} (Off)"

            # Plot the point
            ax.scatter(x_val, y_val, color=color, marker=marker, s=100,
                       label=legend_label if legend_label not in used_labels else None)

            # Add label text above the circle
            player_label = f"{role}_{number}"  # e.g., "WR_1.0"
            ax.annotate(player_label, (x_val, y_val), textcoords="offset points",
                        xytext=(0, 10), ha='center', fontsize=9, color=color)

            # If it's a new legend label, add it to used_labels
            used_labels.add(legend_label)

    # Optional: add a main title with row_index or gameId/playId if present
    if row_index is not None:
        ax.set_title(f"Row index: {row_index}")
    else:
        pass

    ax.legend(loc='upper right', bbox_to_anchor=(1.25, 1.0))
    plt.show()
```
